```python
class doublyNode:
    def __init__(self,elem,next,prev):
        self.elem=elem
        self.next=next #to store next node
        self.prev=prev #to store previous node
```

```python
def createLinkedList(a):
    dh=doublyNode(None,None,None)
    tail=dh
    
    for i in range(len(a)):
        obj=doublyNode(a[i],dh,tail)
        tail.next=obj
        tail=tail.next
        dh.prev=tail
    return dh 

def printLinkedList(dh):
    temp=dh.next
    while temp!=dh:
        print(temp.elem)
        temp=temp.next

head = createLinkedList(np.array([10,15,34,41,56,72]))  
#a=[10,15,34,41,56,72] 
#head = createLinkedList(a)  
printLinkedList(head)

```

```python
def nodeAt(dh,idx):
    temp=dh.next
    c=0
    while temp != dh:
        if c==idx:
            return temp
        c+=1
        temp=temp.next
    return None    #INVALID INDEX
head = createLinkedList(np.array([10,15,34,41,56,72]))  
print(nodeAt(head,2).elem)

```

```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group


class DoublyNode:
    def __init__(self, elem, next=None, prev=None):
        self.elem = elem
        self.next = next
        self.prev = prev


class WRM:
    def __init__(self):
        self.head = DoublyNode(None)  # Dummy head node
        self.size = 0

    def register_patient(self, id, name, age, blood_group):
        patient = Patient(id, name, age, blood_group)
        new_node = DoublyNode(patient)

        if self.size == 0:
            new_node.next = new_node
            new_node.prev = new_node
            self.head.next = new_node
        else:
            last_node = self.head.prev
            new_node.next = self.head.next
            new_node.prev = last_node
            last_node.next = new_node
            self.head.prev = new_node

        self.size += 1
        print("Patient registered successfully.")

    def serve_patient(self):
        if self.size == 0:
            print("No patients in the waiting room.")
            return

        first_patient = self.head.next.elem
        self.head.next = self.head.next.next
        self.head.next.prev = self.head

        self.size -= 1
        print("Now serving:", first_patient.name)

    def cancel_all(self):
        if self.size == 0:
            print("No appointments to cancel.")
            return

        self.head.next = self.head
        self.head.prev = self.head
        self.size = 0
        print("All appointments canceled. Doctor can go to lunch.")

    def can_doctor_go_home(self):
        if self.size == 0:
            print("Yes, the doctor can go home.")
        else:
            print("No, there are patients waiting.")

    def show_all_patients(self):
        if self.size == 0:
            print("No patients in the waiting room.")
            return

        current_node = self.head.next
        print("Waiting patients:")
        while current_node != self.head:
            print(current_node.elem.id)
            current_node = current_node.next


# Tester code
wrm = WRM()

while True:
    print("\nWaiting Room Management System")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        wrm.register_patient(id, name, age, blood_group)
    elif choice == 2:
        wrm.serve_patient()
    elif choice == 3:
        wrm.show_all_patients()
    elif choice == 4:
        wrm.can_doctor_go_home()
    elif choice == 5:
        wrm.cancel_all()
    elif choice == 6:
        break
    else:
        print("Invalid choice. Please try again.")

```

```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group
        self.next = None
        self.prev = None

class WRM:
    def __init__(self):
        self.head = Patient(None, None, None, None)
        self.head.next = self.head
        self.head.prev = self.head

    def register_patient(self, id, name, age, blood_group):
        new_patient = Patient(id, name, age, blood_group)
        new_patient.next = self.head
        new_patient.prev = self.head.prev
        self.head.prev.next = new_patient
        self.head.prev = new_patient
        print("Patient registered successfully.")

    def serve_patient(self):
        if self.head.next == self.head:
            print("No patients in the waiting room.")
            return

        served_patient = self.head.next
        self.head.next = served_patient.next
        served_patient.next.prev = self.head
        print("Serving patient:", served_patient.name)

    def cancel_all(self):
        if self.head.next == self.head:
            print("No appointments to cancel.")
            return

        self.head.next = self.head
        self.head.prev = self.head
        print("All appointments canceled.")

    def can_doctor_go_home(self):
        if self.head.next == self.head:
            print("Yes, doctor can go home.")
        else:
            print("No, patients are still waiting.")

    def show_all_patients(self):
        if self.head.next == self.head:
            print("No patients in the waiting room.")
            return

        current_patient = self.head.next
        while current_patient != self.head:
            print("Patient ID:", current_patient.id)
            current_patient = current_patient.next

# Tester code
wrm = WRM()

while True:
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        wrm.register_patient(id, name, age, blood_group)

    elif choice == 2:
        wrm.serve_patient()

    elif choice == 3:
        wrm.show_all_patients()

    elif choice == 4:
        wrm.can_doctor_go_home()

    elif choice == 5:
        wrm.cancel_all()

    elif choice == 6:
        break

    print()

```
