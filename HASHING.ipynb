```python
class Node:
    def __init__(self,elem,next):
        self.elem=elem
        self.next=next
```

```python
data={'CSE111' : 300,
      'CSE220' : 250,
      'CSE320' : 150,
      'CSE250' : 170,
      'CSE370' : 200}
```

```python
hash_table= [None]*4 #HASH TABLE CREATION
```

```python
def hash_function(key):
    hash_value=ord(key[-1]) + ord(key[-2]) + ord(key[-3])  #CSE111 IT TAKES ONLY [111]
    hash_value=hash_value % 4 
    return hash_value
```

```python
# hash table creation
for k,v in data.items():
    index=hash_function(k)
    if hash_table[index] == None:
        temp=Node((k,v) , None)
        hash_table[index]=temp
    else:
        temp=Node((k,v),None)
        temp.next=hash_table[index]
        hash_table[index]=temp
```

```python
def print_linked_list(head):
    if head is None: return
    while head is not None:
        print(head.elem, end=" -->")
        head=head.next
    print(None)
```

```python
def search_key(key):
    index=hash_function(key)
    if hash_table[index] == None:
        return 'key not found'
    else:
        head=hash_table[index]
        while head is not None:
            if head.elem[0] == key:
                return head.elem[1]
            head=head.next
        return 'key not found'
print(search_key('CSE220'))
```
