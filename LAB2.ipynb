```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

```python
#task 1
def walk_zigzag(floor):
  #To Do
  row,col=floor.shape
  # print(row)
  for i in range(row):
    if i%2 == 0:
      for j in range(0,col,2):
        print(floor[i,j],end=' ')
    else:
      for j in range(col-1,-1,-2):
        print(floor[i,j],end=' ')
    print()



floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

```python
#task 4
def show_knight_move(knight):
  #To Do
  row,col=knight

  a=np.zeros((8,8),dtype=int)
  a[row,col]=33

  if col-2>=0 and row-1>=0:
    a[row-1,col-2]=3

  if col-2>=0 and row+1<=7:
    a[row+1,col-2]=3

  if col+2<=7 and row-1>=0:
     a[row-1,col+2]=3

  if col+2<=7 and row+1<=7:
     a[row+1,col+2]=3


  if col-1>=0 and row-2>=0:
    a[row-2,col-1]=3

  if col-1>=0 and row+2<=7:
    a[row+2,col-1]=3

  if col+1<=7 and row-2>=0:
     a[row-2,col+1]=3

  if col+1<=7 and row+2<=7:
     a[row+2,col+1]=3

  return a

knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
